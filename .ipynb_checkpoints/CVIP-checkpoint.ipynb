```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread("cat.jpg")
img2 = cv2.imread("dog.jpg")
if img1.shape != img2.shape:
    img2 = cv2.resize(img2,(img1.shape[1],img1.shape[0]))
    
out = cv2.absdiff(img1,img2)
m1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
m2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
diff = cv2.cvtColor(out,cv2.COLOR_BGR2RGB)
plt.figure(figsize = (12,6))

plt.subplot(1, 3, 1)
plt.title("Input Image 1")
plt.imshow(m1)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Input Image 2")
plt.imshow(m2)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Difference Output")
plt.imshow(diff)
plt.axis('off')
```

```python
#Ex-2 Question-1 Preprocessing
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("dog.jpg",cv2.IMREAD_GRAYSCALE)

min_val = np.min(img)
max_val = np.max(img)

contrasted = ((img- min_val)*(255/(max_val-min_val))).astype(np.uint8)

kernel= np.ones((5,5),np.float32)/25
filtered= cv2.filter2D(contrasted,-1,kernel)

plt.figure(figsize = (12,6))
plt.subplot(2,3,1)
plt.title("Normal")
plt.imshow(img,cmap = "gray")
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Normal")
plt.imshow(contrasted,cmap = "gray")
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Normal")
plt.imshow(filtered,cmap = "gray")
plt.axis('off')

#Histogram

plt.subplot(2,3,4)
plt.title("Hist-Normal")
plt.hist(img.ravel(),bins = 256,range=[0,256],color="blue")

plt.subplot(2,3,5)
plt.title("Hist-contrastyed")
plt.hist(contrasted.ravel(),bins = 256,range=[0,256],color="blue")

plt.subplot(2,3,6)
plt.title("Hist-filtered")
plt.hist(filtered.ravel(),bins = 256,range=[0,256],color="blue")

plt.tight_layout()
plt.show()
```

```python
#Ex-3 Question-1 image transformation

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("cat.jpg")
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

scale_x = 0.7
scale_y = 0.7

scaled_img = cv2.resize(rgb_img,None,fx = scale_x,fy = scale_y,interpolation=cv2.INTER_LINEAR)

(h,w) = rgb_img.shape[:2]
center = (w//2,h//2)
angle = 45
scale = 1
mrot = cv2.getRotationMatrix2D(center,angle,scale)
rotated_img = cv2.warpAffine(rgb_img,mrot,(w,h))

shear_x = 0.3
shear_y = 0
sheared_mat = np.float32([[1,shear_x,0],[shear_y,1,0]])
sheared_img = cv2.warpAffine(rgb_img,sheared_mat,(w,h))

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(rgb_img)
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Scaled Image")
plt.imshow(scaled_img)
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Rotated Image")
plt.imshow(rotated_img)
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Sheared Image")
plt.imshow(sheared_img)
plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
#Ex-7 question-1
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image and convert to grayscale
img = cv2.imread("tom.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Gaussian blur
grayscaled = cv2.GaussianBlur(img_gray, (5,5), 1.4)

plt.figure(figsize=(12,6))
plt.subplot(2,3,1)
plt.imshow(grayscaled, cmap='gray')
plt.title("Gaussian Blurred")
plt.axis('off')

# Gradient calculation
grad_x = cv2.Sobel(grayscaled, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(grayscaled, cv2.CV_64F, 0, 1, ksize=3)

grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)
grad_magnitude = (grad_magnitude / grad_magnitude.max() * 255).astype(np.uint8)
grad_direction = np.arctan2(grad_y, grad_x)

plt.subplot(2,3,2)
plt.imshow(grad_magnitude, cmap='gray')
plt.title("Gradient Magnitude")
plt.axis('off')

# Non-Maximum Suppression
M, N = grad_magnitude.shape
nms = np.zeros_like(grad_magnitude, dtype=np.uint8)
direction = ((grad_direction + 22.5) // 45) % 4

for i in range(1, M-1):
    for j in range(1, N-1):
        if direction[i,j] == 0:  # 0 deg
            neighbors = [grad_magnitude[i, j+1], grad_magnitude[i, j-1]]
        elif direction[i,j] == 1:  # 45 deg
            neighbors = [grad_magnitude[i+1, j-1], grad_magnitude[i-1, j+1]]
        elif direction[i,j] == 2:  # 90 deg
            neighbors = [grad_magnitude[i+1, j], grad_magnitude[i-1, j]]
        else:  # 135 deg
            neighbors = [grad_magnitude[i-1, j-1], grad_magnitude[i+1, j+1]]
        
        if grad_magnitude[i,j] >= max(neighbors):
            nms[i,j] = grad_magnitude[i,j]

plt.subplot(2,3,3)
plt.imshow(nms, cmap='gray')
plt.title("NMS Output")
plt.axis('off')

# Canny Edge Detection
edges = cv2.Canny(grayscaled, 50, 150)
plt.subplot(2,3,4)
plt.imshow(edges, cmap='gray')
plt.title("Edges (Canny)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Ex-7 question-2

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("tom.jpg",cv2.IMREAD_GRAYSCALE)

def seeding(img,seed,treshhold = 10):
    h,w = img.shape
    segmented = np.zeros_like(img,dtype = np.uint8)
    visited = np.zeros_like(img,dtype = bool)
    seed_list = [seed]
    seed_value = int(img[seed])
    while seed_list:
        x,y = seed_list.pop(0)
        if(visited[x,y]):
            continue
        visited[x,y] = True
        if abs(int(img[x,y]) - seed_value) <= treshhold: 
            segmented[x,y] = 255
            for dx, dy in [(-1,0), (1,0), (0,-1), (0,1)]:
                nx,ny = dx+x,dy+y
                if 0 <= nx < h and 0 <= ny < w and not visited[nx, ny]:
                     seed_list.append((nx, ny))
    return segmented

seed= (120,110)
segmented = seeding(img, seed, treshhold=15)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(segmented, cmap='gray')
plt.title("Region Growing Segmentation")
plt.axis('off')
plt.show()
```
